```python
# Fundamental_Astro_Hw2 By Yang Shuyun 22344020

import numpy as np
import matplotlib.pyplot as plt
from astropy import constants as const

r_sun = const.R_sun
l_sun = const.L_sun
sigma_sb = const.sigma_sb

def luminosity(mass_multiple):
    return l_sun * (mass_multiple) ** 4

def radius(mass_multiple):
    return r_sun * (mass_multiple) ** 0.57

def temperature(mass_multiple):
    return ((luminosity(mass_multiple) / (4 * np.pi * (radius(mass_multiple) ** 2) * sigma_sb)) ** 0.25)

def wavelength_max(t):
    return 0.002897 * 10**9 / t
```

```python
# question1

question1 = np.array([1.1, 1.2, 1.5, 2.0])
t = temperature(question1)
wl_max = wavelength_max(t)

for i in [0, 1, 2, 3]:
    print('The surface temperature of a star, which is ' + str(question1[i]) + f" times the mass of the Sun, is denoted as {t[i]:.2f}, and it corresponds to a wavelength {wl_max[i].value:.2f} nm at which the intensity of radiation peaks.")
    print(' ')
```

```python
# question2

masses = np.linspace(1, 2, 10)
luminosities = luminosity(masses)
temperatures = temperature(masses)

masses_fine = np.linspace(1, 2, 100)
luminosities_fine = luminosity(masses_fine)
temperatures_fine = temperature(masses_fine)

plt.figure(figsize=(10, 6))

sc = plt.scatter(temperatures, luminosities, c=masses, cmap='viridis', s=200, edgecolor='black', label='Sampled Stars')
plt.plot(temperatures_fine, luminosities_fine, 'r-', label='Smoothed Star Line')

plt.title('Hertzsprung-Russell Diagram with Mass Color Bar')
plt.xlabel('Surface Temperature (K)')
plt.ylabel('Luminosity (W/Hz)')

plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()

plt.colorbar(sc, label='Mass (in Solar Masses)')

plt.show()
```

```python
# question3

def calculate_luminosity_ratio(t1, t2, r1, r2):
    # 
    luminosity_ratio = ((t1 / t2) ** 4) * ((r1 / r2) ** 2)
    return luminosity_ratio

t_earth = 300
t_venus = 730
r_earth = 6400000
r_venus = 6073000


# 
luminosity_ratio = calculate_luminosity_ratio(t_venus, t_earth, r_venus, r_earth)
print("The luminosity of Venuss surface is {:.2f} times that of Earths surface.".format(luminosity_ratio))
```
