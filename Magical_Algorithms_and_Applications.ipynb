```python
## Savitzky-Golay滤波器

import numpy as np
from scipy.signal import savgol_filter
import matplotlib.pyplot as plt

# 创建一些示例数据
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x) + np.random.normal(0, 0.1, size=x.shape)

# 使用Savitzky-Golay滤波器平滑数据
window_size = 11  # 滤波器窗口大小
poly_order = 3    # 多项式拟合阶数

smoothed_y = savgol_filter(y, window_size, poly_order)

# 绘制原始数据和平滑后的数据
plt.plot(x, y, label='Original Data')
plt.plot(x, smoothed_y, label='Smoothed Data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
## 基于Numpy.convolve实现滑动平均滤波

import numpy as np
import matplotlib.pyplot as plt

# 生成一些示例数据
np.random.seed(0)
data = np.random.randn(100)

# 定义滑动窗口大小
window_size = 5

# 定义滑动平均滤波器的权重
weights = np.ones(window_size) / window_size

# 使用np.convolve函数进行滑动平均滤波
smoothed_data = np.convolve(data, weights, mode='valid')  # 保持输出与输入相同的大小

# 绘制原始数据和滤波后的数据
plt.plot(data, label='Original Data')
plt.plot(smoothed_data, label='Smoothed Data', linewidth=2)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Moving Average Filter')
plt.legend()
plt.show()
```
